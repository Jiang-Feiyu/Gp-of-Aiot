### **1. ÈááÊ†∑ÁéáÁöÑÂÆö‰πâ**
**ÈááÊ†∑Áéá (Sampling Rate)** ÊòØ‰ø°Âè∑Â§ÑÁêÜ‰∏≠‰∏Ä‰∏™Ê†∏ÂøÉÊ¶ÇÂøµÔºåÂÆÉË°®Á§∫ÊØèÁßíÂØπËøûÁª≠‰ø°Âè∑ÈááÊ†∑ÁöÑÊ¨°Êï∞ÔºåÂçï‰Ωç‰∏∫ **Ëµ´ÂÖπ (Hz)**„ÄÇÈááÊ†∑ÁéáÂèØ‰ª•ÁêÜËß£‰∏∫ÊØèÁßíÈááÈõÜÁöÑÊï∞ÊçÆÁÇπÊï∞„ÄÇ

‰∏æ‰æãÔºö
- Â¶ÇÊûúÈááÊ†∑Áéá‰∏∫ **100 Hz**ÔºåË°®Á§∫ÊØèÁßíÈááÊ†∑ **100 ‰∏™ÁÇπ**„ÄÇ
- Â¶ÇÊûúÈááÊ†∑Áéá‰∏∫ **1 kHzÔºà1000 HzÔºâ**ÔºåË°®Á§∫ÊØèÁßíÈááÊ†∑ **1000 ‰∏™ÁÇπ**„ÄÇ

---

### **2. ‰∏ªÂë®Êúü‰∏é‰∏ªÈ¢ëÁéáÁöÑÂÖ≥Á≥ª**
Âú®Âë®ÊúüÊÄß‰ø°Âè∑‰∏≠Ôºå**‰∏ªÂë®Êúü** (Dominant Period) Âíå **‰∏ªÈ¢ëÁéá** (Dominant Frequency) ÊòØ‰∫íÁõ∏ÂÖ≥ËÅîÁöÑÔºåÂÆÉ‰ª¨ÁöÑÂÖ≥Á≥ª‰∏∫Ôºö

\[
\text{È¢ëÁéá (Hz)} = \frac{1}{\text{Âë®Êúü (Áßí)}}
\]
- **Âë®Êúü** Ë°®Á§∫‰ø°Âè∑ÈáçÂ§ç‰∏ÄÊ¨°ÊâÄÈúÄÁöÑÊó∂Èó¥„ÄÇ
- **È¢ëÁéá** Ë°®Á§∫‰ø°Âè∑Âú® 1 ÁßíÂÜÖÁöÑÈáçÂ§çÊ¨°Êï∞„ÄÇ

---

### **3. Êï∞ÊçÆÈááÊ†∑ÁöÑÊÉÖÂÜµ‰∏ãÂ¶Ç‰ΩïËÆ°ÁÆóÈ¢ëÁéá**
Âú®Á¶ªÊï£‰ø°Âè∑‰∏≠ÔºàÂç≥Êï∞Â≠ó‰ø°Âè∑ÔºâÔºåÊó∂Èó¥ÊòØÈÄöËøáÈááÊ†∑ÁÇπÈó¥ÈöîÈöêÂºèË°®Á§∫ÁöÑ„ÄÇ

#### **ÈááÊ†∑ÁÇπ‰∏éÊó∂Èó¥ÁöÑÂÖ≥Á≥ª**
- Â¶ÇÊûúÈááÊ†∑Áéá‰∏∫ \( fs \)ÔºàÂçï‰ΩçÔºöHzÔºâÔºåÊØè‰∏™ÈááÊ†∑ÁÇπ‰πãÈó¥ÁöÑÊó∂Èó¥Èó¥Èöî‰∏∫Ôºö
  \[
  \Delta t = \frac{1}{fs}
  \]
  ÂÖ∂‰∏≠Ôºö
  - \( \Delta t \) Ë°®Á§∫Áõ∏ÈÇªÈááÊ†∑ÁÇπ‰πãÈó¥ÁöÑÊó∂Èó¥Èó¥ÈöîÔºàÁßíÔºâ„ÄÇ
  - \( fs \) Ë°®Á§∫ÊØèÁßíÁöÑÈááÊ†∑ÁÇπÊï∞„ÄÇ

- Â¶ÇÊûú‰∏ªÂë®ÊúüÊòØ \( P \) ‰∏™ÈááÊ†∑ÁÇπÔºå‰∏ªÂë®ÊúüÊâÄÂØπÂ∫îÁöÑÊó∂Èó¥‰∏∫Ôºö
  \[
  T = P \cdot \Delta t = \frac{P}{fs}
  \]

#### **ËÆ°ÁÆóÈ¢ëÁéá**
Ê†πÊçÆÂë®Êúü \( T \)ÔºåÈ¢ëÁéá \( f \) ÂèØ‰ª•Ë°®Á§∫‰∏∫Ôºö
\[
f = \frac{1}{T} = \frac{fs}{P}
\]

---

### **4. ‰∏∫‰ªÄ‰πàËÉΩÂ§üÁî®ÂÖ¨ÂºèËÆ°ÁÆó‰∏ªÈ¢ëÁéá**
‰ª£Á†Å‰∏≠ÁöÑÂÖ¨ÂºèÔºö
```python
dominant_frequency = fs / dominant_period
```
ÂèØ‰ª•Ê≠£Á°ÆËÆ°ÁÆó‰∏ªÈ¢ëÁéáÁöÑÂéüÂõ†Âú®‰∫éÔºö
- **`fs`** ÊòØÈááÊ†∑ÁéáÔºå‰ª£Ë°®ÊØèÁßíÈááÊ†∑ÁöÑÁÇπÊï∞„ÄÇ
- **`dominant_period`** ÊòØ‰∏ªÂë®ÊúüÔºåË°®Á§∫‰ø°Âè∑Âú®Êó∂Âüü‰∏≠ÈáçÂ§ç‰∏ÄÊ¨°ÊâÄÈúÄÁöÑÈááÊ†∑ÁÇπÊï∞„ÄÇ

‰ª£ÂÖ•Ôºö
\[
f = \frac{fs}{P}
\]
- **ÂàÜÊØç**Ôºö`P` Ë°®Á§∫‰∏ªÂë®ÊúüÁöÑÈááÊ†∑ÁÇπÊï∞„ÄÇ
- **ÂàÜÂ≠ê**Ôºö`fs` Ë°®Á§∫ÊØèÁßíÁöÑÈááÊ†∑ÁÇπÊï∞„ÄÇ
- ÁªìÊûúÊòØ‰∏ªÈ¢ëÁéá \( f \) ÁöÑÂçï‰Ωç‰∏∫ **Ëµ´ÂÖπ (Hz)**„ÄÇ

---

### **5. Á§∫‰æãËØ¥Êòé**
ÂÅáËÆæÔºö
- ÈááÊ†∑Áéá \( fs = 100 \, \text{Hz} \)ÔºàÊØèÁßíÈááÊ†∑ 100 ‰∏™ÁÇπÔºâ„ÄÇ
- ‰∏ªÂë®ÊúüÂØπÂ∫îÁöÑÈááÊ†∑ÁÇπÊï∞ \( P = 25 \) ‰∏™ÁÇπ„ÄÇ

Ê†πÊçÆÂÖ¨ÂºèÔºö
\[
f = \frac{fs}{P} = \frac{100}{25} = 4 \, \text{Hz}
\]
ËøôË°®Á§∫‰ø°Âè∑ÁöÑ‰∏ªÈ¢ëÁéáÊòØ **4 Hz**ÔºåÂç≥‰ø°Âè∑Âú® 1 ÁßíÂÜÖÈáçÂ§ç‰∫Ü 4 Ê¨°„ÄÇ

---

### **6. ÂëºÂê∏È¢ëÁéáÁöÑËÆ°ÁÆó**
ÂëºÂê∏È¢ëÁéáÈÄöÂ∏∏‰ª• **ÊØèÂàÜÈíüÁöÑÂëºÂê∏Ê¨°Êï∞ (BPM)** Ë°®Á§∫„ÄÇ‰ªéÈ¢ëÁéá \( f \)ÔºàÂçï‰ΩçÔºöHzÔºâËΩ¨Êç¢‰∏∫ BPM ÁöÑÂÖ¨ÂºèÊòØÔºö
\[
\text{BPM} = f \cdot 60
\]

ÁªìÂêà‰ª•‰∏äÔºö
- Â¶ÇÊûúÈááÊ†∑Áéá \( fs = 100 \, \text{Hz} \)Ôºå‰∏ªÂë®Êúü \( P = 25 \)Ôºö
  \[
  f = \frac{fs}{P} = \frac{100}{25} = 4 \, \text{Hz}
  \]
  ËΩ¨Êç¢‰∏∫ BPMÔºö
  \[
  \text{BPM} = 4 \cdot 60 = 240 \, \text{BPM}
  \]

---

### **7. ÊÄªÁªì**

- **ÈááÊ†∑Áéá (fs)**ÔºöË°®Á§∫ÊØèÁßíÈááÈõÜÁöÑÁÇπÊï∞ÔºåÁî®‰∫éÂ∞ÜÁ¶ªÊï£‰ø°Âè∑ÁöÑÈááÊ†∑ÁÇπËΩ¨Êç¢‰∏∫ÂÆûÈôÖÊó∂Èó¥„ÄÇ
- **ÂÖ¨Âºè \( f = fs / P \)**ÔºöÂà©Áî®ÈááÊ†∑ÁéáÂíå‰∏ªÂë®ÊúüÁöÑÈááÊ†∑ÁÇπÊï∞ÔºåËÆ°ÁÆó‰ø°Âè∑ÁöÑ‰∏ªÈ¢ëÁéá„ÄÇ
- **‰ø°Âè∑Âë®ÊúüÂíåÈ¢ëÁéáÁöÑÂÖ≥Á≥ª**Ôºö‰∏ªÂë®ÊúüË∂äÁü≠ÔºàÈááÊ†∑ÁÇπË∂äÂ∞ëÔºâÔºå‰∏ªÈ¢ëÁéáË∂äÈ´òÔºõ‰∏ªÂë®ÊúüË∂äÈïøÔºå‰∏ªÈ¢ëÁéáË∂ä‰Ωé„ÄÇ

Â¶ÇÊûúËøòÊúâÂÖ∂‰ªñÈóÆÈ¢òÔºåÊ¨¢ËøéÁªßÁª≠ÊèêÈóÆÔºÅ üòä

````python
import pandas as pd
import numpy as np
from scipy import signal as sp_signal
import matplotlib.pyplot as plt
import json


# **1. Âä†ËΩΩÊï∞ÊçÆ**
def load_csi_data(file_path):
    """
    Âä†ËΩΩÂíåËß£Êûê CSI Êï∞ÊçÆÈõÜ
    :param file_path: CSV Êñá‰ª∂Ë∑ØÂæÑ
    :return: DataFrameÔºåÂåÖÂê´ÂÖÉÊï∞ÊçÆÂíåËß£ÊûêÂêéÁöÑ CSI Êï∞ÊçÆ
    """
    # Âä†ËΩΩ CSV Êñá‰ª∂
    df = pd.read_csv(file_path)

    # ÊèêÂèñ CSI Êï∞ÊçÆÂàóÂπ∂Ëß£Êûê
    def parse_csi_data(csi_string):
        try:
            # Â∞ÜÂ≠óÁ¨¶‰∏≤ÂΩ¢ÂºèÁöÑ CSI Êï∞ÊçÆËß£Êûê‰∏∫ NumPy Êï∞ÁªÑ
            return np.array(json.loads(csi_string))
        except (json.JSONDecodeError, TypeError) as e:
            print(f"Error parsing CSI data: {e}")
            return np.nan  # Â¶ÇÊûúËß£ÊûêÂ§±Ë¥•ÔºåËøîÂõû NaN

    # Ëß£Êûê `data` Âàó
    df['csi_array'] = df['data'].apply(parse_csi_data)

    # Âà†Èô§Êó†ÊïàË°å
    df = df.dropna(subset=['csi_array']).reset_index(drop=True)
    return df


# **2. È¢ÑÂ§ÑÁêÜÊï∞ÊçÆ**
def preprocess_csi_data(csi_data, metric="amplitude"):
    """
    ‰ªé CSI Êï∞ÊçÆ‰∏≠ÊèêÂèñÂπÖÂ∫¶ÊàñÁõ∏‰ΩçÔºåÂπ∂ËøõË°åÈ¢ÑÂ§ÑÁêÜ
    :param csi_data: DataFrame ‰∏≠ÁöÑ 'csi_array' ÂàóÔºåÂåÖÂê´ÊØè‰∏™ÈááÊ†∑ÁÇπÁöÑ CSI Êï∞ÊçÆ
    :param metric: ÈÄâÊã© "amplitude" Êàñ "phase" ‰Ωú‰∏∫ÊèêÂèñÊåáÊ†á
    :return: ÊèêÂèñÂêéÁöÑ‰ø°Âè∑Êï∞ÁªÑÔºåÂΩ¢Áä∂‰∏∫ (N, M)ÔºåN ‰∏∫ÈááÊ†∑ÁÇπÊï∞ÔºåM ‰∏∫Â≠êËΩΩÊ≥¢Êï∞Èáè
    """
    num_subcarriers = len(csi_data.iloc[0]) // 2  # ÊØè‰∏™ÈááÊ†∑ÁÇπÁöÑÂ≠êËΩΩÊ≥¢Êï∞Èáè
    processed_data = []

    for csi in csi_data:
        subcarrier_values = []
        for subcarrier_idx in range(num_subcarriers):
            # ÊèêÂèñÂ≠êËΩΩÊ≥¢ÁöÑËôöÈÉ®ÂíåÂÆûÈÉ®
            imaginary = csi[subcarrier_idx * 2]
            real = csi[subcarrier_idx * 2 + 1]

            # ËÆ°ÁÆóÂπÖÂ∫¶ÊàñÁõ∏‰Ωç
            if metric == "amplitude":
                subcarrier_values.append(np.sqrt(imaginary ** 2 + real ** 2))  # ËÆ°ÁÆóÂπÖÂ∫¶
            elif metric == "phase":
                subcarrier_values.append(np.arctan2(imaginary, real))  # ËÆ°ÁÆóÁõ∏‰Ωç

        processed_data.append(subcarrier_values)
    return np.array(processed_data)  # ËøîÂõûÂΩ¢Áä∂‰∏∫ (N, M) ÁöÑ‰∫åÁª¥Êï∞ÁªÑ


# **3. ÂëºÂê∏È¢ëÁéá‰º∞ËÆ°ÔºöËá™Áõ∏ÂÖ≥ÂáΩÊï∞ (ACF) Âä†ÊùÉÂπ≥ÂùáÔºåÂπ∂Â∫îÁî®‰∏≠ÂÄºÊª§Ê≥¢**
def estimate_breathing_rate_acf_weighted(signal_matrix, fs):
    """
    ‰ΩøÁî®Âä†ÊùÉÂπ≥ÂùáÁöÑËá™Áõ∏ÂÖ≥ÂáΩÊï∞ (ACF) ‰ªé‰ø°Âè∑‰∏≠‰º∞ËÆ°ÂëºÂê∏È¢ëÁéá
    :param signal_matrix: ‰ø°Âè∑Áü©ÈòµÔºåÂΩ¢Áä∂‰∏∫ (N, M)ÔºåN ‰∏∫ÈááÊ†∑ÁÇπÊï∞ÔºåM ‰∏∫Â≠êËΩΩÊ≥¢Êï∞Èáè
    :param fs: ÈááÊ†∑Áéá (Hz)
    :return: ÂëºÂê∏È¢ëÁéá (BPM)
    """
    num_subcarriers = signal_matrix.shape[1]
    # ÁÆÄÂçïÁ§∫‰æãÔºöÊ†πÊçÆÂ≠êËΩΩÊ≥¢ÊñπÂ∑ÆÂàÜÈÖçÊùÉÈáç
    variances = np.var(signal_matrix, axis=0)
    weights = 1 / (variances + 1e-8)  # ÈÅøÂÖçÈô§Èõ∂
    weights = weights / np.sum(weights)  # ÂΩí‰∏ÄÂåñÊùÉÈáç

    avg_acf = np.zeros(signal_matrix.shape[0])

    # ËÆ°ÁÆóÊØè‰∏™Â≠êËΩΩÊ≥¢ÁöÑ ACF Âπ∂Âä†ÊùÉÊ±ÇÂíåÔºåÂêåÊó∂Â∫îÁî®‰∏≠ÂÄºÊª§Ê≥¢
    for subcarrier in range(num_subcarriers):
        signal = signal_matrix[:, subcarrier]
        acf = np.correlate(signal, signal, mode='full')
        acf = acf[len(acf) // 2:]  # Âè™‰øùÁïôÊ≠£ÂçäÈÉ®ÂàÜ
        acf = sp_signal.medfilt(acf, kernel_size=3)  # Â∫îÁî®‰∏≠ÂÄºÊª§Ê≥¢
        avg_acf += weights[subcarrier] * acf

    # ÊâæÂà∞‰∏ªÂë®Êúü
    peaks, _ = sp_signal.find_peaks(avg_acf)
    if len(peaks) == 0:
        return 0  # Ê≤°ÊúâÊâæÂà∞Â≥∞ÂÄºÔºåËøîÂõû 0

    # ÊâæÂà∞Á¨¨‰∏Ä‰∏™Ê≥¢Â≥∞Âà∞ÂºÄÂßãÂèñÊ†∑ÁöÑÊó∂Èó¥Â∑ÆÔºàÊ†∑Êú¨ÁÇπÊï∞Ôºâ
    dominant_period = peaks[0]

    # ËΩ¨Êç¢‰∏∫ÂëºÂê∏È¢ëÁéá (Hz)
    breathing_rate_hz = fs / dominant_period
    breathing_rate_bpm = breathing_rate_hz * 60  # ËΩ¨Êç¢‰∏∫ BPM
    return breathing_rate_bpm


# **4. Â∏¶ÈÄöÊª§Ê≥¢Âô®**
def apply_bandpass_filter(signal, fs, lowcut=0.15, highcut=0.5, order=4):
    """
    ÂØπ‰ø°Âè∑ËøõË°åÂ∏¶ÈÄöÊª§Ê≥¢
    :param signal: ËæìÂÖ•‰ø°Âè∑
    :param fs: ÈááÊ†∑Áéá (Hz)
    :param lowcut: Â∏¶ÈÄöÊª§Ê≥¢Âô®ÁöÑ‰ΩéÈ¢ëÊà™Ê≠¢È¢ëÁéá (Hz)
    :param highcut: Â∏¶ÈÄöÊª§Ê≥¢Âô®ÁöÑÈ´òÈ¢ëÊà™Ê≠¢È¢ëÁéá (Hz)
    :param order: Â∏¶ÈÄöÊª§Ê≥¢Âô®ÁöÑÈò∂Êï∞
    :return: Êª§Ê≥¢ÂêéÁöÑ‰ø°Âè∑
    """
    nyquist = fs / 2
    if highcut >= nyquist or lowcut <= 0:
        raise ValueError("Cutoff frequencies must be in the range (0, Nyquist frequency).")
    b, a = sp_signal.butter(order, [lowcut / nyquist, highcut / nyquist], btype='bandpass')
    return sp_signal.filtfilt(b, a, signal, axis=0)


# **5. ÂÆûÊó∂ÁõëÊµãÂëºÂê∏È¢ëÁéá**
def monitor_breathing_rate(csi_df, window_size_seconds, step_size_seconds, metric="amplitude"):
    """
    ÂÆûÊó∂ÁõëÊµãÂëºÂê∏È¢ëÁéá
    :param csi_df: ÂåÖÂê´ CSI Êï∞ÊçÆÁöÑ DataFrameÔºåÈúÄÂåÖÂê´ 'timestamp' Âíå 'csi_array' Âàó
    :param window_size_seconds: Á™óÂè£Â§ßÂ∞èÔºàÁßíÔºâ
    :param step_size_seconds: ÊªëÂä®Ê≠•ÈïøÔºàÁßíÔºâ
    :param metric: ÊèêÂèñÁöÑÊåáÊ†áÔºå"amplitude" Êàñ "phase"
    :return: ÊØè‰∏™Á™óÂè£ÁöÑÂëºÂê∏È¢ëÁéáÂíåÊó∂Èó¥ËåÉÂõ¥ÁªìÊûúÂàóË°®
    """
    timestamps = pd.to_datetime(csi_df['timestamp'])
    csi_array = csi_df['csi_array']

    # Âä®ÊÄÅËÆ°ÁÆóÈááÊ†∑Áéá
    time_deltas = np.diff(timestamps.astype('int64') / 1e9)
    avg_sampling_interval = np.mean(time_deltas)
    fs = 1 / avg_sampling_interval  # ÈááÊ†∑Áéá (Hz)
    print(f"Sampling rate: {fs:.2f} Hz")

    # ËÆ°ÁÆóÁ™óÂè£Â§ßÂ∞èÂíåÊ≠•ÈïøÔºàÂçï‰ΩçÔºöÊ†∑Êú¨Êï∞Ôºâ
    window_size_samples = int(window_size_seconds * fs)
    step_size_samples = int(step_size_seconds * fs)

    # ÂàùÂßãÂåñÁªìÊûúÂàóË°®
    breathing_rates = []

    for start_idx in range(0, len(csi_array) - window_size_samples + 1, step_size_samples):
        # ÊèêÂèñÁ™óÂè£Êï∞ÊçÆ
        window_signal = np.array(csi_array[start_idx:start_idx + window_size_samples].tolist())

        # Êª§Ê≥¢
        try:
            filtered_signal = apply_bandpass_filter(window_signal, fs)
        except Exception as e:
            print(f"Filtering error at index {start_idx}: {e}")
            continue

        # ÂëºÂê∏È¢ëÁéá‰º∞ËÆ°
        try:
            rate = estimate_breathing_rate_acf_weighted(filtered_signal, fs)
        except Exception as e:
            print(f"ACF error at index {start_idx}: {e}")
            continue

        # ‰øùÂ≠òÁªìÊûú
        breathing_rates.append({
            "start_time": timestamps.iloc[start_idx],
            "end_time": timestamps.iloc[start_idx + window_size_samples - 1],
            "breathing_rate": rate
        })

    return breathing_rates


# **6. ÂèØËßÜÂåñÂëºÂê∏È¢ëÁéáÂèòÂåñ**
def plot_breathing_rate(breathing_rates):
    """
    ÁªòÂà∂ÂëºÂê∏È¢ëÁéáÈöèÊó∂Èó¥ÂèòÂåñÁöÑÊõ≤Á∫ø
    :param breathing_rates: ÂåÖÂê´ÂëºÂê∏È¢ëÁéáÂíåÊó∂Èó¥ËåÉÂõ¥ÁöÑÁªìÊûúÂàóË°®
    """
    times = [result['start_time'] for result in breathing_rates]
    rates = [result['breathing_rate'] for result in breathing_rates]

    plt.figure(figsize=(10, 6))
    plt.plot(times, rates, marker='o', linestyle='-', color='b')
    plt.title("Breathing Rate Over Time")
    plt.xlabel("Time")
    plt.ylabel("Breathing Rate (BPM)")
    plt.grid(True)
    plt.show()


# ËÆ°ÁÆó MAE ÁöÑÂáΩÊï∞
def calculate_mae(predicted, actual):
    """
    ËÆ°ÁÆóÈ¢ÑÊµãÂÄºÂíåÁúüÂÆûÂÄº‰πãÈó¥ÁöÑÂπ≥ÂùáÁªùÂØπËØØÂ∑Æ
    :param predicted: È¢ÑÊµãÁöÑÂëºÂê∏È¢ëÁéáÂàóË°®
    :param actual: ÁúüÂÆûÁöÑÂëºÂê∏È¢ëÁéáÂàóË°®
    :return: MAE ÂÄº
    """
    predicted = np.array(predicted)
    actual = np.array(actual)
    return np.mean(np.abs(predicted - actual))


# **7. ‰∏ªÁ®ãÂ∫è**
if __name__ == "__main__":
    file_path = "D:/Espressif/frameworks/esp-idf-v5.4/comp7310_2025_group_project/benchmark/breathing_rate/evaluation/CSI20250227_191018.csv"
    csi_df = load_csi_data(file_path)

    # ÂëºÂê∏È¢ëÁéáÁõëÊµã
    breathing_rates = monitor_breathing_rate(csi_df, window_size_seconds=15, step_size_seconds=1, metric="amplitude")

    # ËæìÂá∫ÁªìÊûú
    for rate_info in breathing_rates:
        print(f"Start Time: {rate_info['start_time']}, End Time: {rate_info['end_time']}, Breathing Rate: {rate_info['breathing_rate']} BPM")

    # Âä†ËΩΩÁúüÂÆûÂëºÂê∏È¢ëÁéáÊï∞ÊçÆ
    true_breathing_file = "D:/Espressif/frameworks/esp-idf-v5.4/comp7310_2025_group_project/benchmark/breathing_rate/evaluation/gt_20250227_191018.csv"
    true_breathing_df = pd.read_csv(true_breathing_file)
    # Â§ÑÁêÜ time ÈáçÂ§çÂÄºÔºåÂèñÊúÄÂêé‰∏Ä‰∏™ÂÄºÂØπÂ∫îÁöÑ bpm
    true_breathing_df = true_breathing_df.drop_duplicates(subset='time', keep='last')
    # ÂÅáËÆæÊó∂Èó¥ËåÉÂõ¥ËÉΩÂåπÈÖç‰∏äÔºåÊèêÂèñÁúüÂÆûÁöÑ bpm
    actual_breathing_rates = true_breathing_df['bpm'].tolist()

    # ÊèêÂèñÈ¢ÑÊµãÁöÑÂëºÂê∏È¢ëÁéá
    predicted_breathing_rates = [rate_info['breathing_rate'] for rate_info in breathing_rates]

    # Á°Æ‰øùÈ¢ÑÊµãÂÄºÂíåÁúüÂÆûÂÄºÈïøÂ∫¶‰∏ÄËá¥
    min_length = min(len(predicted_breathing_rates), len(actual_breathing_rates))
    predicted_breathing_rates = predicted_breathing_rates[:min_length]
    actual_breathing_rates = actual_breathing_rates[:min_length]

    # ËÆ°ÁÆó MAE
    mae = calculate_mae(predicted_breathing_rates, actual_breathing_rates)
    print(f"Mean Absolute Error (MAE): {mae:.2f} BPM")

    # ÂëºÂê∏È¢ëÁéáÂèØËßÜÂåñ
    plot_breathing_rate(breathing_rates)
    
````
